```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/02-analysis_data/analysis_data.csv")
```

```python
# plot number of buildings by year built
fig = px.histogram(df, x="YEAR_BUILT", nbins=25, title="Distribution of Year Built")
fig.show()
```

```python
# plot average number of stories by year built
fig = px.histogram(df, x="YEAR_BUILT", y="CONFIRMED_STOREYS", histfunc="avg", title="Average Number of Stories by Year Built")
fig.show()
```

```python
# plot total number of units by year built
fig = px.histogram(df, x="YEAR_BUILT", y="CONFIRMED_UNITS", nbins=25, title="Total Units by Year Built", histfunc="sum")
fig.show()
```

```python
# plot the average score by year built
fig = px.line(df.groupby("YEAR_BUILT")["SCORE"].mean().reset_index(), x="YEAR_BUILT", y="SCORE", title="Average Score by Year Built")
fig.show()
```
